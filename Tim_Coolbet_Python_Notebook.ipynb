```python
# ganti nama kolom
df_japan_renamed = df_japan.rename(columns={
    'REF_AREA': 'Country_Region',
    'MEASURE': 'Measure_Code',
    'Measure': 'Measure_Description',
    'FREQ': 'Observation_Frequency',
    'TIME_PERIOD': 'Year',
    'OBS_VALUE': 'Life_Expectancy',
    'Unit of measure': 'Unit_of_Measurement',
    'STRUCTURE': 'Data_Type',
    'TIMELINESS': 'Timeliness_Description',
    'DECIMALS': 'Number_of_Decimals'
})
df_japan_renamed.head()
```

```python
# Melakukan pivot
df_pivot = selected_named.pivot_table(
    index=['Country_Region', 'Year'],
    columns='Measure_Description',
    values='Life_Expectancy',
    aggfunc=lambda x: x.iloc[0]  # Mengambil nilai pertama dari data asli
)

# Mengatur nama kolom
df_pivot.columns.name = None  # Menghapus nama kolom yang dihasilkan oleh pivot_table
df_pivot.reset_index(inplace=True)  # Mengubah index menjadi kolom biasa

df_pivot
```

```python
df_pivot = df_pivot.drop(columns=['Country_Region'])
print(df_pivot)
```

### 1. Mencari missing value dan mengatasi masalah missing value, jika ada

```python
df_pivot.isnull().sum()
```

```python
# Menghitung jumlah missing values per kolom
n_missing = df_pivot.isnull().sum()

# Menghitung proporsi missing values per kolom
missing_proportion = n_missing / len(df_pivot)

# Menampilkan hasil untuk setiap kolom
for column in df_pivot.columns:
    print(f"Column: {column}")
    print(f"  Proportion of dataset missing: {missing_proportion[column]:.2%}")
```

```python
# Mengatur Year sbg index
df_pivot.set_index("Year", inplace=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat subplots
fig, ax = plt.subplots(figsize=[10, 5])

# Plot 'Value added' dan 'Labour tax revenue' dengan garis untuk menunjukkan data
ax.plot(df_pivot.index, df_pivot['Value added'], label='Value added', color='b', marker='o', linestyle='-', markersize=6)
ax.plot(df_pivot.index, df_pivot['Labour tax revenue'], label='Labour tax revenue', color='g', marker='o', linestyle='-', markersize=6)

# Menambahkan judul dan label sumbu
ax.set_title("Labour Tax Revenue and Value Added with Missing Data")
ax.set_ylabel("Value")
ax.set_xlabel("Year")

# Menampilkan semua tahun di sumbu x
ax.set_xticks(df_pivot.index)
ax.set_xticklabels(df_pivot.index, rotation=45)

# Menambahkan legenda
ax.legend(title="Variables and Missing Values")

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python

# Cek nilai yang hilang dalam 'Value added' dan 'Labour tax revenue'
print("Missing value sebelum imputasi:")
print(df_pivot[['Value added', 'Labour tax revenue']].isna().sum())

# Imputasi data hilang
# Menggunakan forward fill (ffill) dan backward fill (bfill) untuk mengisi data yang hilang
df_pivot['Value added'] = df_pivot['Value added'].fillna(method='ffill')  # Forward fill
df_pivot['Labour tax revenue'] = df_pivot['Labour tax revenue'].fillna(method='ffill')  # Forward fill

# Jika masih ada nilai yang hilang setelah forward fill, gunakan backward fill
df_pivot['Value added'] = df_pivot['Value added'].fillna(method='bfill')  # Backward fill
df_pivot['Labour tax revenue'] = df_pivot['Labour tax revenue'].fillna(method='bfill')  # Backward fill

# Cek missing value setelah imputasi
print("Missing value setelah imputasi:")
print(df_pivot[['Value added', 'Labour tax revenue']].isna().sum())

# Menampilkan DataFrame setelah imputasi
print("DataFrame setelah imputasi:")
print(df_pivot[['Value added', 'Labour tax revenue']])
```

```python
# Mengecek kembali apakah masih terdapat Missing Value
df_pivot.isnull().sum()
```

Interpretasi: Untuk mengatasi **missing value** pada data **Labour tax revenue** dan **Value added** di Jepang adalah dengan menggunakan Interpolation backward dan forward filling karena karena data kami merupakan data deret waktu (time series) jadi penggunaan Interpolation Backward dan Forward filling memungkinkan kita untuk mengisi nilai yang hilang dengan cara yang logis dan sesuai dengan konteks urutan waktu. Teknik-teknik ini berusaha untuk mengisi kekosongan dengan nilai yang "berkelanjutan" atau "terkait" dengan data sebelumnya atau sesudahnya.
Setelah di lakukan pengisian dengan Forward dan Backward filling maka missing value sudah teratasi dan hasilnya dapat di lihat pada output diatas.

##### Uji Normalitas

```python
#Import library
from scipy import stats
import pandas as pd

# Menghitung uji normalitas Shapiro-Wilk untuk setiap fitur kecuali dan 'Country_Region'
normality_results = {}

# Exclude 'Year' and 'Country_Region' columns from the normality test
for column in df_pivot.select_dtypes(include=['number']):
    stat, p_value = stats.shapiro(df_pivot[column])
    normality_results[column] = {'Statistic': stat, 'p-value': p_value}

# Menampilkan hasil uji normalitas
normality_df = pd.DataFrame(normality_results).T
normality_df['Normal'] = normality_df['p-value'] > 0.05
print(normality_df)
```

**Insight:**

**1. Statistik dan p-value:**
* **Statistik:** Nilai yang dihitung dari data yang menunjukkan seberapa jauh data tersebut menyimpang dari distribusi normal.
* **p-value:** Nilai yang menunjukkan probabilitas bahwa hasil yang diamati terjadi secara kebetulan jika hipotesis nol benar. Umumnya, p-value di bawah 0.05 atau 0.01 dianggap signifikan, yang menunjukkan bahwa data tidak mengikuti distribusi normal.

**2. Normalitas:**
* Semua variabel dalam daftar menunjukkan bahwa mereka tidak berdistribusi normal (semua memiliki nilai "False" untuk kolom "Normal"). Ini berarti bahwa data dari variabel-variabel ini tidak mengikuti distribusi normal, yang dapat mempengaruhi analisis lebih lanjut seperti regresi atau pengujian hipotesis yang mengasumsikan normalitas.

##### Uji Normalitas

```python
from scipy import stats
import pandas as pd

# Menghitung uji normalitas Shapiro-Wilk untuk setiap fitur kecuali 'Year' dan 'Country_Region'
normality_results = {}

# Exclude 'Year' and 'Country_Region' columns from the normality test
for column in df_pivot_asia.select_dtypes(include=['number']).drop('Year', axis=1).columns:
    stat, p_value = stats.shapiro(df_pivot_asia[column])
    normality_results[column] = {'Statistic': stat, 'p-value': p_value}

# Menampilkan hasil uji normalitas
normality_df = pd.DataFrame(normality_results).T
normality_df['Normal'] = normality_df['p-value'] > 0.05
print(normality_df)
```

**Insight:**

**1. Statistik dan p-value:**
* **Statistik:** Nilai yang dihitung dari data yang menunjukkan seberapa jauh data tersebut menyimpang dari distribusi normal.
* **p-value:** Nilai yang menunjukkan probabilitas bahwa hasil yang diamati terjadi secara kebetulan jika hipotesis nol benar. Umumnya, p-value di bawah 0.05 atau 0.01 dianggap signifikan, yang menunjukkan bahwa data tidak mengikuti distribusi normal.

**2. Normalitas:**
* Semua variabel dalam daftar menunjukkan bahwa mereka tidak berdistribusi normal (semua memiliki nilai "False" untuk kolom "Normal"). Ini berarti bahwa data dari variabel-variabel ini tidak mengikuti distribusi normal, yang dapat mempengaruhi analisis lebih lanjut seperti regresi atau pengujian hipotesis yang mengasumsikan normalitas.

**Variabel yang Diperiksa:**
* **Female population:** Terdapat nilai statistik 0.684540 dan p-value yang sangat kecil (2.545784e-19), menunjukkan bahwa data ini tidak berdistribusi normal.

* **GDP deflator:** Nilai statistik 0.935957 dengan p-value 8.120306e-08 juga menunjukkan non-normalitas.

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import adfuller
```

```python
ts2 = df_pivot.copy()
```

Dalam tampilan data diatas tidak ada pola yang jelas dari data tersebut. Penggunaan moldel ARIMA akan cocok untuk data tersebut, karena model ini fleksibel untuk menangkap pola tren, season, atau perulangan.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Dekomposisi (mode 'additive' atau 'multiplicative' sesuai data)
decomposition = seasonal_decompose(ts2['Real GDP'], model='additive', period=12)  # Sesuaikan period dengan data

# Plot hasil dekomposisi
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(ts2['Real GDP'], label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# 1. Uji stasionaritas (ADF Test)
def check_stationarity(series):
    result = adfuller(series)
    print("Hasil Uji ADF:")
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print(f"Critical Values: {result[4]}")
    if result[1] < 0.05:
        print("Data stasioner pada tingkat signifikansi 5%.")
    else:
        print("Data tidak stasioner pada tingkat signifikansi 5%.")

check_stationarity(ts2['Real GDP'])
```

###Cek Stasionaritas

```python
# 2. Differencing jika data tidak stasioner
df_diff = ts2['Real GDP'].diff().dropna()
check_stationarity(df_diff)
```

Setelah diffrencing pertama kali, data sudah mulai stasioner

```python
# Visualisasi differencing
plt.figure(figsize=(10, 5))
plt.plot(ts2.index[1:], df_diff, marker='o', label='Differenced Data')
plt.title('Data Setelah Differencing')
plt.xlabel('Tahun')
plt.ylabel('Perubahan Nilai')
plt.legend()
plt.show()
```

###Cek ACF dan PACF

```python
# 3. Pemilihan parameter ARIMA (p, d, q)
# Gunakan ACF dan PACF untuk menentukan nilai p dan q
plot_acf(df_diff, lags=32)
plt.title('ACF (Autocorrelation)')
plt.show()

plot_pacf(df_diff, lags=11)
plt.title('PACF (Partial Autocorrelation)')
plt.show()
```

Korelasi data yang ditampilkan menunjukan adanya beberapa lag yang diluar interval, ini akan berpengaruh pada penentuan model ARIMA yang akan digunakan. berdasarkan data diatas, tidak terjadi adanya indikasi lag yang banyak sehingga data kemungkinan besar tidak memiliki seasonalitas

###Model ARIMA

Untuk menentukan nilai p, q, dan r kita harus melihat dulu korelasi data dan proses diffrencing sebelumnya.<br>
p : didapat dari pada urutan ke berapa lag data yang berada diluar interval di grafik PACF<br>
d : didapat dari jumlah proses diffrencing yang dilakukan sampai data stationer<br>
q : didapat dari urutan ke berapa lag data yang berada diluar interval di grafik ACF<br><br>

sehingga bisa didapatkan nilai :
p = 11, d = 1, q = 11

```python
p, d, q = 11, 1, 11

# 4. Pelatihan model ARIMA
model = ARIMA(ts2['Real GDP'], order=(p, d, q))
model_fit = model.fit()

# Ringkasan model
print(model_fit.summary())
```

###Grafik Pelatihan

```python
# 5. Visualisasi hasil pelatihan
plt.figure(figsize=(10, 5))
plt.plot(ts2.index, ts2['Real GDP'], label='Data Asli')
plt.plot(ts2.index, model_fit.fittedvalues, label='Hasil Pelatihan', linestyle='--')
plt.title('Hasil Pelatihan ARIMA')
plt.xlabel('Tahun')
plt.ylabel('Nilai')
plt.legend()
plt.show()
```

Terlihat model data pelatihan memeberikan hasil yang bagus, dikarenakan hasil pelatihan mampu mendekati data Asli. Namun untuk untuk analisa pola masih diperlukan percobaan lebih lanjut

###Grafik Prediksi

```python
# 6. Prediksi ke depan
forecast = model_fit.get_forecast(steps=5)
forecast_index = range(ts2.index[-1] + 1, ts2.index[-1] + 6)
forecast_values = forecast.predicted_mean
conf_int = forecast.conf_int()
```

```python
# Visualisasi prediksi
plt.figure(figsize=(10, 5))
plt.plot(ts2.index, ts2['Real GDP'], label='Data Asli')
plt.plot(forecast_index, forecast_values, marker='o', color='red', label='Prediksi')
plt.fill_between(forecast_index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='pink', alpha=0.3, label='Interval Kepercayaan')
plt.title('Prediksi dengan ARIMA')
plt.xlabel('Tahun')
plt.ylabel('Real GDP')
plt.legend()
plt.show()
```
